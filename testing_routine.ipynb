```python
test_filename = 'test.csv'

import sklearn
import pandas as pd
import numpy as np

from sklearn.pipeline import Pipeline
from sklearn.experimental import enable_iterative_imputer
from sklearn.ensemble import IsolationForest, GradientBoostingClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
from imblearn.over_sampling import RandomOverSampler, SMOTE
from sklearn.preprocessing import OneHotEncoder

from joblib import load
from sklearn.metrics import accuracy_score

np.random.seed = 123
```

```python
#lettura del test set
data = pd.read_csv(test_filename, sep=",")
x_ts = data.iloc[:,:-1]
y_ts = data.iloc[:,-1]

#variabili categoriche
categorical_columns = [col for col in x_ts.columns if x_ts[col].dtype == "object"]

#variabili numeriche
numerical_columns = [col for col in x_ts.columns if x_ts[col].dtype in ["int64", "float64"]]

#preprocessing per dati numerici
numerical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", RobustScaler())
])

#preprocessing per i dati categorici
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore"))
])

#caricamento preprocessor serializzato
preprocessor = load('preprocessing.joblib')
x_ts = preprocessor.transform(x_ts)

#caricamento classificatore serializzato
classificator = load('classificator.joblib')
print(accuracy_score(y_ts, classificator.predict(x_ts)))

```
